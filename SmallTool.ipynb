```python
class SmallTool(object):
    
    def Rows2Columns(self,df):
        
        data = df[['advertId','targetApp']].fillna('None')
        
        data.columns = ['condition','object']
        
        data['object_process'] = data['object'].apply(lambda x: re.compile(r"'appId':(.*?),").findall(x))
        
        data['object_list'] = data['object_process'].apply(lambda x: str(x).strip('[]').split(','))
        
        new_list = []
        
        for index,dataframe in data.iterrows():
            for i in dataframe['object_list']:
                new_list.append([dataframe['condition'],i])
        
        result = pd.DataFrame(new_list).drop_duplicates()
        
        return result.to_excel('C:/Users/DELL/Desktop/targetApp.xlsx')
    
    def Columns2Rows(self,df):
        pass
        
    def AderCpcCtrRegulation(self,df):
        
        data = df.copy()
        
        data['dtstr'] = data['dt'].apply(lambda x: str(x)[:10])
        
#         data['time'] = data['dt'].apply(lambda x: str(x)).str.cat([data['hour'].apply(lambda x: str(x)),data['minute']],sep='-')

        data_1 = data[data['rank-2']!=0]
        data_1 = data_1.sort_values(by=['dt','hour','minute'])
    
        dflist = []
        
        for i in pd.unique(data_1['ader']):
            data_single = data_1[data_1.ader==i]
            data_1_pt = data_single.pivot_table(index=['dtstr','hour','minute'],values=['xh','jfdj','fq'],aggfunc='sum').reset_index()
            data_1_pt['cpc'] = data_1_pt['xh']/data_1_pt['jfdj']
            data_1_pt['ctr'] = data_1_pt['jfdj']/data_1_pt['fq']
#             data_1_pt['xhdiff'] = data_1_pt.groupby('dtstr')['xh'].apply(lambda x:x.diff(1))
#             data_1_pt['cpcdiff'] = data_1_pt.groupby('dtstr')['cpc'].apply(lambda x:x.diff(1))
            data_1_pt['fqdiff'] = data_1_pt.groupby('dtstr')['fq'].apply(lambda x:x.diff(1))
            data_1_pt['fqshift'] = data_1_pt.groupby('dtstr')['fq'].apply(lambda x:x.shift(1))
            data_1_pt['fqratio'] = data_1_pt['fqdiff']/data_1_pt['fqshift']
            data_1_pt['ader'] = i
            data_x = data_1_pt[(data_1_pt.fqratio>=-0.1)&(data_1_pt.fqratio<=0.1)]
            data_x['corr'] = data_x[['xh','cpc']].corr().loc['xh','cpc']
            dflist.append(data_x)
            
        result = pd.concat(dflist,axis=0)    
            
        return result.to_excel('C:/Users/DELL/Desktop/rank-2.xlsx')
```
